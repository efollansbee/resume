```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
#Define Constants
Po = 101325 #Pa
g = 9.80665 #m/s^2
Rd = 287.0967 #J/K/kg
```

```python
#Create a list of temperature to height
z1 = [0, 3000, 6000, 9000, 11000, 20000, 23000, 26000, 29000, 32000, 35000, 38000, 
      41000, 44000, 47000, 51000, 54000, 57000, 60000, 63000, 66000, 69000, 71000]
T1 = [288.15, 268.650, 249.150, 229.654, 216.65, 216.650, 219.650, 222.650, 225.650, 
      228.65, 237.050, 245.450, 253.850, 262.250, 270.65, 270.65, 262.250, 253.850, 245.450, 237.050, 228.650, 220.250, 214.65]

fig, ax = plt.subplots()
ax.plot(T1, z1);
ax.set_ylabel('Height, z (m)');
ax.set_xlabel('Pressure, P (Pa)');
ax.set_title('Pressure vs Height');
```

```python
# Solving for each T(z)
import scipy.integrate as integrate

P1 = []

def get_p(T1, z1):
    pres1 = Po*np.exp((-g/Rd)*expint(T1, z1))
    return(pres1)


def integrand(z1, T1, P1):
    return(1 / T1)

def expint(T1, z1):
    I = integrate.quad(integrand, 0, z1, args=(T1, z1))[0]
    return(I)

for i in range(len(T1)):
    P1.append(get_p(T1[i],z1[i]))

P1
```

```python
fig, ax = plt.subplots()
ax.plot(P1, z1);
ax.set_xlabel('P(z)');
ax.set_ylabel('z');
ax.set_title('P(z) vs. z');
```

```python
fig, ax = plt.subplots();
ax.plot(P1, z1);
ax.set_xscale('log');
ax.set_xlabel('log P(z)');
ax.set_ylabel('z');
ax.set_title('log P(z) vs. z');
```

### Explain why log P(z) vs. z is a straight line

##### This is because when you take the log of an exponential function, it becomes a linear function.

```python
#for an isothermal atmosphere

z2 = [0, 3000, 6000, 9000, 11000, 20000, 23000, 26000, 29000, 32000, 35000, 38000, 
      41000, 44000, 47000, 51000, 54000, 57000, 60000, 63000, 66000, 69000, 71000]
T2 = [288.15, 268.650, 249.150, 229.654, 216.65, 216.650, 219.650, 222.650, 225.650, 
      228.65, 237.050, 245.450, 253.850, 262.250, 270.65, 270.65, 262.250, 253.850, 245.450, 237.050, 228.650, 220.250, 214.65]

avg_T = np.mean(T2)
avg_T
```

```python
P3 = []

def get_p(z2):
    pres = Po*np.exp(-g/(Rd*avg_T)*z2)
    print(pres)
    return(pres)

for i in range(len(T2)):
    P3.append(get_p(z2[i]))
```

```python
fig, ax = plt.subplots()
ax.plot(P3, z2);
ax.plot(P1, z2);
ax.set_xlabel('P(z)');
ax.set_ylabel('z');
ax.set_title('Isothermal Atmosphere: P(z) vs. z');
```

```python
fig, ax = plt.subplots()
ax.plot(P3, z2);
ax.plot(P1, z2);
ax.set_xscale('log');
ax.set_xlabel('log P(z)');
ax.set_ylabel('z');
ax.set_title('Isothermal Atmosphere: log P(z) vs. z');
```

#### You can see that for the isothermal atmosphere (blue line), the line is perfectly linear showing that the temperature is the same throughout the atmosphere. As compared to the orange line which is close to linear, but not perfect showing the temperature changes and inconsistencies throughout the atmosphere.
